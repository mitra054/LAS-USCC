```python
import numpy as np
from scipy import linalg
from pyscf import gto, scf, lib, mcscf
from mrh.my_pyscf.mcscf.lasscf_o0 import LASSCF
from mrh.exploratory.citools import fockspace
from mrh.exploratory.unitary_cc import lasuccsd as lasuccsd
from mrh.exploratory.unitary_cc.uccsd_sym0 import get_uccs_op
from mrh.exploratory.unitary_cc.uccsd_sym0 import get_uccsd_op
import itertools
```

```python
xyz = '''H 0.0 0.0 0.0;
            H 1.0 0.0 0.0;
            H 0.2 1.6 0.1;
            H 1.159166 1.3 -0.1'''
mol = gto.M (atom = xyz, basis = 'sto-3g', output='h4_sto3g.log.py',
    verbose=0)
mf = scf.RHF (mol).run ()
ref = mcscf.CASSCF (mf, 2, 4).run () # = FCI
las = LASSCF (mf, (1,1), (2,2), spin_sub=(1,1))
las.verbose=1
frag_atom_list = ((0,1),(2,3))
mo_loc = las.localize_init_guess (frag_atom_list, mf.mo_coeff)
las.kernel (mo_loc)
```

```python
def save_to_xyz_file(mol_pyscf, filename="molecule_challenge.xyz"):
    with open(filename, 'w') as f:
        f.write(str(mol_pyscf.natm) + '\n')  # Number of atoms
        f.write('\n')  # Comment line
        for atom in mol_pyscf._atom:
            element, coords = atom[0], atom[1]
            f.write('{} {:.6f} {:.6f} {:.6f}\n'.format(element, *coords))

# Save molecule to XYZ format
save_to_xyz_file(mol)
```

```python
mc = mcscf.CASCI (mf, 4, 4)
mc.mo_coeff = las.mo_coeff
mc.fcisolver = lasuccsd.FCISolver (mol)
mc.fcisolver.norb_f = [2,2] # Number of orbitals per fragment
mc.verbose=5
mc.kernel()
print("34--------------------")

print ("FCI energy:      {:.9f}".format (ref.e_tot))
print ("LASSCF energy:   {:.9f}".format (las.e_tot))
print ("LASUCCSD energy: {:.9f}\n".format (mc.e_tot))

# mc.ci has to be a full CI vector ndarray. In order to access more
# information, a "LASUCCTrialState" object is cached on the fcisolver
psi = mc.fcisolver.psi

x = psi.x # Amplitude vector that solves the BFGS problem 
# The trailing elements of x are all zero because the fragment CI vectors
# are updated at the end of a calculation.
# print("Print x",x)
h1, h0 = mc.get_h1eff ()
h2 = mc.get_h2eff ()
h = [h0, h1, h2] # 2nd-quantized CAS Hamiltonian
energy, gradient = psi.e_de (x, h)
print ("Recomputing LASUCC total energy with cached objective function")
print ("LASUCCSD energy: {:.9f}".format (energy))
print ("|gradient| = {:.3e}".format (linalg.norm (gradient)))
print ("If that seems too high to you, consider: BFGS sucks.\n")


all_g, g, gen_indices, a_idxs_new, i_idxs_new, num_a_idxs, num_i_idxs = psi.get_grad_t1 (x, h, epsilon=2.0)

print ('Number of a_idxs in input',num_a_idxs)
print ('Number of i_idxs in input',num_i_idxs)
#exc_list = psi.get_excitation_list()
# Get the CI coefficients from psi
ci_f = psi.ci_f
norb_f = psi.norb_f
print("--------------------")

mol = gto.M (atom = xyz, basis = 'sto-3g', output='h4_sto3g2.log.py',
    verbose=lib.logger.DEBUG)
mf = scf.RHF (mol).run ()
ref = mcscf.CASSCF (mf, 4, 4).run () # = FCI
```

```python
# ... [the rest of your imports]
import matplotlib.pyplot as plt

# ... [the rest of your setup code]

epsilons = np.logspace(-4.2, -4.7, num=30)  # Generating epsilon values on a logarithmic scale
energies = []
lengths = []  # Store lengths of a_idxs_new for each epsilon

for epsilon in epsilons:
    all_g, g, gen_indices, a_idxs_new, i_idxs_new, num_a_idxs, num_i_idxs = psi.get_grad_t1(x, h, epsilon=epsilon)
    lengths.append(len(a_idxs_new))
    
    # Rest of your code that computes the energy for this epsilon
    mol = gto.M(atom=xyz, basis='sto-3g', output='h4_sto3g2.log.py',
                verbose=lib.logger.DEBUG)
    mf = scf.RHF(mol).run()
    mc_uscc = mcscf.CASCI(mf, 4, 4)
    mc_uscc.mo_coeff = las.mo_coeff
    mc_uscc.fcisolver = lasuccsd.FCISolver2(mol, a_idxs_new, i_idxs_new)
    mc_uscc.fcisolver.norb_f = [2,2]
    mc_uscc.verbose = 5
    mc_uscc.kernel()
    energies.append(mc_uscc.e_tot)

# Use LAS-UCCSD as reference
lasuccsd_energy = mc.e_tot
reference_energy = lasuccsd_energy
relative_energies = [(e - reference_energy) * 627.509 for e in energies]
print("Relative energies",relative_energies)

# Your provided energy values
fci_energy = ref.e_tot
lasscf_energy = las.e_tot

relative_fci = (fci_energy - reference_energy) * 627.509 
relative_lasscf = (lasscf_energy - reference_energy) * 627.509 

# ... [rest of your code before plotting]

# Plotting
fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(epsilons, relative_energies, '-o', color='gray', label="Delta E (kcal/mol)")
ax.axhline(relative_fci, color='red', linestyle='solid', label="E_FCI - E_ref")
#ax.axhline(relative_lasscf, color='green', linestyle='--', label="E_LASSCF - E_ref")
ax.axhline(0, color='blue', linestyle='--', label="E_LASUCCSD (Reference)")

major_ticks = np.logspace(-6.2, -3.8, 6)  # Adjusted range for major ticks
minor_ticks = np.logspace(-6.2, -3.8, 21)  # Adjusted range for minor ticks

ax.set_xticks(major_ticks)
ax.set_xticks(minor_ticks, minor=True)
ax.get_xaxis().set_major_formatter(plt.ScalarFormatter())

ax.set_xscale('log')
ax.set_xlabel('Epsilon', fontsize=14)
ax.set_ylabel('Delta E', fontsize=14)
ax.invert_xaxis()  # Flip the direction of the x-axis
ax.grid(True, which="both", ls="--", c='0.65')

# Set tick label fontsize
ax.tick_params(axis='both', which='major', labelsize=14)
ax.tick_params(axis='both', which='minor', labelsize=14)  # This will set the minor tick label size but it's often not necessary since minor ticks usually don't have labels.

# Add lengths as text annotations
for i, (epsilon, length) in enumerate(zip(epsilons, lengths)):
    ax.text(epsilon, relative_energies[i], f"{length}", ha='center', va='bottom', fontsize=10)
    print(relative_energies[i])

# Adjust the legend position using (x,y) tuple
ax.legend(loc=(0.5, 0.6), fontsize=14)
ax.set_title('Delta E (E - E_ref) vs Epsilon', fontsize=16)
plt.figtext(0.67, 0.81, "Total number of parameters = 142", fontsize=14, ha='center')

# Save the figure as PNG
plt.savefig("energy_vs_epsilon_close_USCC.png", dpi=300)

# Show the plot
plt.show()











```
