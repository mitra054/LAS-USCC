```python
import numpy as np
from scipy import linalg
from pyscf import gto, scf, lib, mcscf
from mrh.my_pyscf.mcscf.lasscf_o0 import LASSCF
from mrh.exploratory.citools import fockspace
from mrh.exploratory.unitary_cc import lasuccsd as lasuccsd
from mrh.exploratory.unitary_cc.uccsd_sym0 import get_uccs_op
from mrh.exploratory.unitary_cc.uccsd_sym0 import get_uccsd_op
import itertools
import pickle
```

```python
xyz = '''H 0.0 0.0 0.0;
            H 1.0 0.0 0.0;
            H 0.2 1.6 0.1;
            H 1.159166 1.3 -0.1'''
mol = gto.M (atom = xyz, basis = 'sto-3g', output='h4_sto3g.log.py',
    verbose=0)
mf = scf.RHF (mol).run ()
ref = mcscf.CASSCF (mf, 4, 4).run () # = FCI
las = LASSCF (mf, (2,2), (2,2), spin_sub=(1,1))
las.verbose=1
frag_atom_list = ((0,1),(2,3))
mo_loc = las.localize_init_guess (frag_atom_list, mf.mo_coeff)
las.kernel (mo_loc)
```

```python
def save_to_xyz_file(mol_pyscf, filename="molecule_challenge.xyz"):
    with open(filename, 'w') as f:
        f.write(str(mol_pyscf.natm) + '\n')  # Number of atoms
        f.write('\n')  # Comment line
        for atom in mol_pyscf._atom:
            element, coords = atom[0], atom[1]
            f.write('{} {:.6f} {:.6f} {:.6f}\n'.format(element, *coords))

# Save molecule to XYZ format
save_to_xyz_file(mol)
```

```python
mc = mcscf.CASCI (mf, 4, 4)
mc.mo_coeff = las.mo_coeff
mc.fcisolver = lasuccsd.FCISolver (mol)
mc.fcisolver.norb_f = [2,2] # Number of orbitals per fragment
mc.verbose=5
mc.kernel()
print("34--------------------")

print ("FCI energy:      {:.9f}".format (ref.e_tot))
print ("LASSCF energy:   {:.9f}".format (las.e_tot))
print ("LASUCCSD energy: {:.9f}\n".format (mc.e_tot))

# mc.ci has to be a full CI vector ndarray. In order to access more
# information, a "LASUCCTrialState" object is cached on the fcisolver
psi = mc.fcisolver.psi

x = psi.x # Amplitude vector that solves the BFGS problem 
# The trailing elements of x are all zero because the fragment CI vectors
# are updated at the end of a calculation.
# print("Print x",x)
h1, h0 = mc.get_h1eff ()
h2 = mc.get_h2eff ()
h = [h0, h1, h2] # 2nd-quantized CAS Hamiltonian
energy, gradient = psi.e_de (x, h)
print ("Recomputing LASUCC total energy with cached objective function")
print ("LASUCCSD energy: {:.9f}".format (energy))
print ("|gradient| = {:.3e}".format (linalg.norm (gradient)))
print ("If that seems too high to you, consider: BFGS sucks.\n")


all_g, g, gen_indices, a_idxs_new, i_idxs_new, num_a_idxs, num_i_idxs = psi.get_grad_t1 (x, h, epsilon=2.0)

print ('Number of a_idxs in input',num_a_idxs)
print ('Number of i_idxs in input',num_i_idxs)
#exc_list = psi.get_excitation_list()
# Get the CI coefficients from psi
ci_f = psi.ci_f
norb_f = psi.norb_f
print("--------------------")

mol = gto.M (atom = xyz, basis = 'sto-3g', output='h4_sto3g2.log.py',
    verbose=lib.logger.DEBUG)
mf = scf.RHF (mol).run ()
ref = mcscf.CASSCF (mf, 4, 4).run () # = FCI
```
